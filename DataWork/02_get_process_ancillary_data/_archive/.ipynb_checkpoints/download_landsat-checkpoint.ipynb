```python
// Get Landsat Bands

var getlandsat = function(tile, begin, end, cloud_cover_filter, filename_prefix, country_name){
  // Import Country Data
  //var countries = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017');
  //var country = countries.filter(ee.Filter.eq('country_na', country_name));

  var countries = ee.FeatureCollection('users/robmarty3/pak_grid_200km');
  var country = countries.filter(ee.Filter.eq('id', tile));

  // Import Image Collection
  var l8 = ee.ImageCollection('LANDSAT/LC08/C01/T1_SR');

  var l8_filter = ee.ImageCollection(
          l8.filterBounds(country)
            .filterDate(begin, end)
            .filter(ee.Filter.lt('CLOUD_COVER', cloud_cover_filter))
        );
        
  var l8_filter = ee.ImageCollection(
          l8_filter.filterBounds(country)
            .filterDate(begin, end)
        );   
        
  var l8_median = l8_filter.reduce(ee.Reducer.median()); 
  l8_median = l8_median.clip(country);

  var l8_b1 = l8_median.select('B1_median');
  var l8_b2 = l8_median.select('B2_median');
  var l8_b3 = l8_median.select('B3_median');
  var l8_b4 = l8_median.select('B4_median');
  var l8_b5 = l8_median.select('B5_median');
  var l8_b6 = l8_median.select('B6_median');
  var l8_b7 = l8_median.select('B7_median');
  var l8_b10 = l8_median.select('B10_median');
  var l8_b11 = l8_median.select('B11_median');

  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b1,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b1',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b2,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b2',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b3,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b3',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b4,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b4',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b5,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b5',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b6,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b6',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b7,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b7',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b10,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b10',
  });
  
  Export.image.toDrive({
    folder: "gee_extracts",
    image: l8_b11,
    scale: 30,
    region: country,
    maxPixels: 9000000000,
    description: filename_prefix + 'tile' + tile + '_b11',
  });
  
  return l8_b1;

};


var l8_pak_2014 = getlandsat(1, '2014-01-01', '2014-12-31', 15, 'l8_2014', 'Pakistan');
//var l8_pak_2015 = getlandsat('2015-01-01', '2015-12-31', 15, 'l8_2015', 'Pakistan');
//var l8_pak_2016 = getlandsat('2016-01-01', '2016-12-31', 15, 'l8_2016', 'Pakistan');
//var l8_pak_2017 = getlandsat('2017-01-01', '2017-12-31', 15, 'l8_2017', 'Pakistan');
//var l8_pak_2018 = getlandsat('2018-01-01', '2018-12-31', 15, 'l8_2018', 'Pakistan');
//var l8_pak_2019 = getlandsat('2019-01-01', '2019-12-31', 15, 'l8_2019', 'Pakistan');


Map.addLayer(l8_pak_2014, {}, '2013');


var i;
for (i = 1; i < 3; i++) {
  var l8_pak_2014 = getlandsat(i, '2014-01-01', '2014-12-31', 15, 'l8_2014', 'Pakistan');
}
```
